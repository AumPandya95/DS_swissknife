```python
import seaborn as sns

# reset index
data.reset_index(inplace=True)

# rename the column to year
data.rename(columns={'index': 'year'}, inplace=True)

# melt df into a long format
df2 = data.melt(id_vars='year', var_name='month')

# create a datatime column and set it as index
df2['date'] = pd.to_datetime(df2['year'].astype(str) + '-' + df2['month'])
df2.set_index('date', inplace=True)

# CO2 dataframe
co = df2[df2['month'].isin(['Nov', 'Dec', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct'])].copy()

# sort the index
co.sort_index(inplace=True)
```

### Using pmdarima package

```python
import pmdarima as pmd

def arimamodel(timeseriesarray):
    arima_model = pmd.ARIMA(order=[0,1,1])
    arima_model.fit(timeseriesarray)
    return arima_model
arima_model = arimamodel(np.array(co.value))
arima_model.summary()
```
