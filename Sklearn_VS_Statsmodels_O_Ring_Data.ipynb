```python
# Data
in_data = pd.read_excel("datasets/o_ring_failure.xlsx", engine="openpyxl")
# target = np.array(ring_failure.loc[:, "number_experiencing_thermal_distress"])
# ring_failure.drop(columns=["number_experiencing_thermal_distress"], inplace=True)
# train_array = np.array(ring_failure)
```

```python
Y = in_data['number_experiencing_thermal_distress']
X = in_data.drop(['number_experiencing_thermal_distress'], axis=1)
```

```python
X = sm.add_constant(X)
X
```

```python
#Linear regression with lasso penalty and only positive coefficients
lm_lasso_sk = linear_model.Lasso(alpha=1, fit_intercept=False, positive=True)
lm_lasso_sk.fit(X, Y)
print(lm_lasso_sk.coef_)
```

```python
# StatsModels
#Linear regression with lasso penalty
lm_lasso_sm = sm.OLS(Y, X)
results_lasso = lm_lasso_sm.fit_regularized(method='elastic_net', alpha=1, L1_wt=1.0)
print(results_lasso.params)
```

```python
np.all(lm_lasso_sk.coef_ == results_lasso.params)
```

```python
set_1 = ['SKlearn_'+str(i) for i in range(4)]
set_2 = ['SM_'+str(i) for i in range(4)]
```

```python
out_frame_sk_pos = pd.DataFrame(data=np.zeros((17, 9))
                        , columns = [['alpha'] + set_1 + set_2])

out_frame_sk_pos
```

```python
# Check the results for multipe values of alpha
# sklearn model returns ONLY POSITIVE
counter = 0
for alpha_val in np.arange(0.3, 2, 0.1):
    lm_lasso_sk = linear_model.Lasso(alpha=alpha_val, fit_intercept=False, positive=True)
    lm_lasso_sk.fit(X, Y)

    lm_lasso_sm = sm.OLS(Y, X)
    results_lasso = lm_lasso_sm.fit_regularized(method='elastic_net', alpha=alpha_val, L1_wt=1.0)

    out_frame_sk_pos.iloc[counter, 0] = alpha_val
    out_frame_sk_pos.iloc[counter, 1:5] = lm_lasso_sk.coef_
    out_frame_sk_pos.iloc[counter, 5:9] = list(results_lasso.params)
    counter+=1
```

```python
out_frame_sk_both = pd.DataFrame(data=np.zeros((17, 9))
                        , columns = [['alpha'] + set_1 + set_2])
```

```python
# Check the results for multipe values of alpha
# sklearn model returns positive and negative coefficients
counter = 0
for alpha_val in np.arange(0.3, 2, 0.1):
    lm_lasso_sk = linear_model.Lasso(alpha=alpha_val, fit_intercept=False, positive=False)
    lm_lasso_sk.fit(X, Y)

    lm_lasso_sm = sm.OLS(Y, X)
    results_lasso = lm_lasso_sm.fit_regularized(method='elastic_net', alpha=alpha_val, L1_wt=1.0)

    out_frame_sk_both.iloc[counter, 0] = alpha_val
    out_frame_sk_both.iloc[counter, 1:5] = lm_lasso_sk.coef_
    out_frame_sk_both.iloc[counter, 5:] = list(results_lasso.params)
    counter+=1
```
